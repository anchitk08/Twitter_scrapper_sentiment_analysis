## Animation plotting of real time twitter data

```python
#%pylab
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from itertools import count
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')

x_values = []
y1 = []
y2 = []

index = count()
frame_size = 250

def animate(i):
    data = pd.read_csv('sentiment.csv')
    y1 = data['usa']
    y2 = data['china']
    xx = next(index)
    x_values.append(xx)
    if xx > frame_size:
        x_values.pop(0)
    plt.cla()
    plt.plot(x_values, y1.iloc[x_values], label = 'usa')
    plt.plot(x_values, y2.iloc[x_values], label = 'china')
    plt.legend(loc = 'upper left')


ani = FuncAnimation(plt.gcf(), animate, 200)


plt.tight_layout()
plt.show()
```
